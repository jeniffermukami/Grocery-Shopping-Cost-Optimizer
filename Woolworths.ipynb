```python
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
driver = webdriver.Chrome()
```

```python
item = input('select brand > ')
```

```python
driver.get('https://www.woolworths.co.za/cat?Ntt=' + item + '&Dy=1')
driver.implicitly_wait(60)
```

```python
html_page = driver.page_source #get source code of the page
```

```python
soup = BeautifulSoup(html_page, 'lxml')
products = soup.find_all('div', class_ = 'product-list__item')
for product in products:
    try:
        product_name = product.find('div', class_ = 'range--title product-card__name').text
    except:
        product_name = "Unknown"
    product_price = product.find('strong', class_ = 'price').text.replace(' ', '')
    try:
        promo_original_price = product.find('strong', class_ = 'price--strikethrough price--original').text.replace(' ', '')
    except:
        promo_original_price = None
 
    original_price = promo_original_price
    url = 'https://www.woolworths.co.za'+product.find('a', href=True)['href']
    
    print(f"Product name: {product_name}")
    print(f"Current price: {product_price}")
    print(f"Old price: {original_price}")
    print(f"Item URL: {url}")
    print(' ')
```
